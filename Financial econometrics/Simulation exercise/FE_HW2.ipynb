```python
import pandas as pd
import numpy as np 
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns 

import plotly.express as px
import plotly.figure_factory as ff

%matplotlib inline
```

Part a)

```python
mean = (4+0)/2
variance = (4.0-0)**2/12
```

Part b)

```python
X = np.random.uniform(0, 4, 100)
```

Part c)

```python
avg = X.mean()
```

Part d)

```python
averages = []
for i in range(1000):
  averages.append(np.random.uniform(0, 4, 100).mean())
```

Part e)

```python
# Plot with seaborn

plt.figure(figsize=(16, 6))
sns.histplot(averages)
```

```python
# Plot with plotly

fig = px.histogram(averages)
fig.show()
```

Part f)

```python
averages = np.array(averages)
z_stat = (averages - mean)/(variance/np.sqrt(100))
```

```python
# Plot with seaborn

plt.figure(figsize=(16, 6))
sns.distplot(z_stat, fit = stats.norm)
```

```python
# Plot with plotly

fig2 = ff.create_distplot([z_stat], group_labels=['distplot'], bin_size=0.1, curve_type='normal')
fig2.update_layout(title_text='Distplot with Normal Distribution')
fig2.show()
```

Part g)

```python
z_stat = np.array([])
for n in range (1000):
  averages = np.array([])
  for j in range(1000):
    averages = np.append(averages, np.random.uniform(0, 4, 100).mean())
  z_stat = np.append(z_stat, (averages - mean)/(variance/np.sqrt(100)))
```

```python
# Plot with seaborn

plt.figure(figsize=(16, 6))
sns.distplot(z_stat, fit = stats.norm)
```
