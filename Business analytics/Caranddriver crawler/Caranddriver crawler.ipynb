```python
# import libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
```

```python
# Open webdriver
driver = webdriver.Edge('Edgedriver.exe')
driver.maximize_window()
```

```python
"""Daryafte url marboot be khodro ha"""
# open sedan webpage
Sedan_2020 = 'https://www.caranddriver.com/features/g27227136/best-sedans/'
driver.get(Sedan_2020)
# Get sedan urls
Sedan_cars = driver.find_elements_by_css_selector(r'body > main > div.content-container.listicle-container > div.listicle-body > div.listicle-body-content > div > div.listicle-slide-dek > p:nth-child(2) > a')
Sedan_urls=[i.get_attribute("href")+r'/specs' for i in Sedan_cars]
# open suv webpage
SUV_2020 = 'https://www.caranddriver.com/features/g27257734/best-suvs/'
driver.get(SUV_2020)
# Get suv urls
SUV_cars = driver.find_elements_by_css_selector(r'body > main > div.content-container.listicle-container > div.listicle-body > div.listicle-body-content > div > div.listicle-slide-dek > p:nth-child(2) > a')
SUV_urls=[i.get_attribute("href")+r'/specs' for i in SUV_cars]
# open sport webpage
Sport_2020 = 'https://www.caranddriver.com/features/g27197524/best-sports-cars/'
driver.get(Sport_2020)
# Get sport urls
Sport_cars = driver.find_elements_by_css_selector(r'body > main > div.content-container.listicle-container > div.listicle-body > div.listicle-body-content > div > div.listicle-slide-dek > p:nth-child(2) > a')
Sport_urls=[i.get_attribute("href")+r'/specs' for i in Sport_cars]
# open pickup-trucks webpage
Pickup_Trucks_2020 = 'https://www.caranddriver.com/features/g27242492/best-pickup-trucks/'
driver.get(Pickup_Trucks_2020)
# Get pickup-trucks urls
Pickup_Trucks_cars = driver.find_elements_by_css_selector(r'body > main > div.content-container.listicle-container > div.listicle-body > div.listicle-body-content > div > div.listicle-slide-dek > p:nth-child(2) > a')
Pickup_Trucks_urls=[i.get_attribute("href")+r'/specs' for i in Pickup_Trucks_cars]
# open hatchbacks webpage
Hatchbacks_2020 = 'https://www.caranddriver.com/features/g26594907/best-hatchbacks/'
driver.get(Hatchbacks_2020)
# Get hatchbacks urls
Hatchbacks_cars = driver.find_elements_by_css_selector(r'body > main > div.content-container.listicle-container > div.listicle-body > div.listicle-body-content > div > div.listicle-slide-dek > p:nth-child(2) > a')
Hatchbacks_urls=[i.get_attribute("href")+r'/specs' for i in Hatchbacks_cars]
# open station-wagons webpage
Station_Wagons_2020 = 'https://www.caranddriver.com/features/g26961972/best-station-wagons/'
driver.get(Station_Wagons_2020)
# Get station-wagons urls
Station_Wagons_cars = driver.find_elements_by_css_selector(r'body > main > div.content-container.listicle-container > div.listicle-body > div.listicle-body-content > div > div.listicle-slide-dek > p:nth-child(2) > a')
Station_Wagons_urls=[i.get_attribute("href")+r'/specs' for i in Station_Wagons_cars]
# open coupes webpage
Coupes_2020 = 'https://www.caranddriver.com/features/g27199030/best-coupes/'
driver.get(Coupes_2020)
# Get coupes urls
Coupes_cars = driver.find_elements_by_css_selector(r'body > main > div.content-container.listicle-container > div.listicle-body > div.listicle-body-content > div > div.listicle-slide-dek > p:nth-child(2) > a')
Coupes_urls=[i.get_attribute("href")+r'/specs' for i in Coupes_cars]
# open convertibles webpage
Convertibles_2020 = 'https://www.caranddriver.com/features/g27198629/convertibles/'
driver.get(Convertibles_2020)
# Get convertibles urls
Convertibles_cars = driver.find_elements_by_css_selector(r'body > main > div.content-container.listicle-container > div.listicle-body > div.listicle-body-content > div > div.listicle-slide-dek > p:nth-child(2) > a')
Convertibles_urls=[i.get_attribute("href")+r'/specs' for i in Convertibles_cars]
```

```python
Name=[]
Model=[]
CarType=[]
Price=[]
Engine_Type_and_Required_Fuel=[]
Cylinders=[]
Displacement=[]
Fuel_system=[]
Horsepower=[]
Mpg=[]
Weight=[]
Url=[]
```

```python
def crawler(Cartype, urls):
    for url in urls:
        try:
            driver.get(url)
        except:
            continue
        try:
            Name.append(driver.find_element_by_css_selector(r'body > main > div.content-container.review-container.vehicle-specs > h1').text[:-19])
        except:
            Name.append(np.nan)
        try:
            Model.append(driver.find_element_by_css_selector(r'body > main > div.content-container.review-container.vehicle-specs > div.review-body > div > div.specs-overview > div.overview-text > div').text[:-17])
        except:
            Model.append(np.nan)
        try:
            CarType.append(Cartype)
        except:
            CarType.append(np.nan)
        try:
            Price.append(int((driver.find_element_by_css_selector(r'body > main > div.content-container.review-container.vehicle-specs > div.review-body > div > div.specs-overview > div.overview-price > div.price').text.replace(",","")).replace("$","")))
        except:
            Price.append(np.nan)
        try:
            Engine_Type_and_Required_Fuel.append(driver.find_element_by_css_selector(r"#engine-specifications-content > div:nth-child(2) > div").text)
        except:
            Engine_Type_and_Required_Fuel.append(np.nan)
        try:
            Cylinders.append(Engine_Type_and_Required_Fuel[-1].split(" ")[-1])
        except:
            Cylinders.append(np.nan)
        try:
            Displacement.append(driver.find_element_by_css_selector(r"#engine-specifications-content > div:nth-child(3) > div").text)
        except:
            Displacement.append(np.nan)
        try:
            Fuel_system.append(driver.find_element_by_css_selector(r'#engine-specifications-content > div:nth-child(4) > div').text)
        except:
            Fuel_system.append(np.nan)
        try:
            Horsepower.append(driver.find_element_by_css_selector(r'#engine-specifications-content > div:nth-child(5) > div').text)
        except:
            Horsepower.append(np.nan)
        try:
            temp = driver.find_element_by_css_selector('#fuel\ economy\ and\ range-specifications-content > div:nth-child(5) > div').text
            if "N" in temp:
                temp=driver.find_element_by_css_selector('#fuel\ economy\ and\ range-specifications-content > div:nth-child(4) > div').text
            Mpg.append(temp)
        except:
            Mpg.append(np.nan)
        try:
            Weight.append(int(driver.find_element_by_css_selector(r'#weight\ information-specifications-content > div > div').text))
        except:
            Weight.append(np.nan)
        try:
            Url.append(url)
        except:
            Url.append(np.nan)
```

```python
crawler("Sedan", Sedan_urls)
crawler("SUV", SUV_urls)
crawler("Sport", Sport_urls)
crawler("Pickup Truck", Pickup_Trucks_urls)
crawler("Hatchback", Hatchbacks_urls)
crawler("Station Wagon", Station_Wagons_urls)
crawler("Coupe", Coupes_urls)
crawler("Convertible", Convertibles_urls)
```

```python
Cars = pd.DataFrame({"Name":Name,
              "Model":Model,
              "Type":CarType,
              "Price":Price,
              "Engine Type and Required Fuel":Engine_Type_and_Required_Fuel,
              "Cylinders":Cylinders,
              "Displacement":Displacement,
              "Fuel system":Fuel_system,
              "Horsepower":Horsepower,
              "Mpg":Mpg,
              "Weight":Weight,
              "URL": Url
             })
```

```python
Cars.to_excel("raw cars output.xlsx")
```
